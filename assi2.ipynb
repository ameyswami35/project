```python
import datacompy
import pandas as pd
```

```python
#reading data from excel file
df1=pd.read_csv('flipkart_com-ecommerce_sample.csv')
```

```python
#reading data from excelfile
df2=pd.read_excel('amz_com-ecommerce_sample.xlsx')
```

```python
#writing simple code to compare 2 data having same product name and givinf name to df1 and df2
comparison=datacompy.Compare(df1,df2,join_columns="product_name",df1_name="Flipkart",df2_name="Amazon")
```

```python
#printing comparision table
print(comparision.report())
```
